5.   Discuss the role of random sampling in making inferences about populations.

=>  Random sampling is a method where every individual in a population has an equal chance of being selected in the sample.

it helps ensure that the sample is:

Representative of the whole population

Unbiased, meaning no group is over- or under-represented

1. Reduces Bias
Random selection helps prevent systematic errors that can occur if we select certain individuals more than others.

Example:
If you only survey students in the library about study habits, your sample might overrepresent serious students.

 2. Supports Generalization
With a proper random sample, statistical inferences (like estimating means, proportions, or testing hypotheses) can be confidently extended from the sample to the population.

Example:
You survey 200 randomly chosen people to estimate the average height of adults in a city. If the sample is truly random, your estimate should be pretty close to the actual average.

 3. Enables Valid Use of Statistical Techniques
Many statistical methods (like confidence intervals and hypothesis tests) assume randomness. If your sample isn't random, these methods can give misleading results.

 4. Allows for Estimation of Error
Random sampling allows us to calculate sampling error, like standard error and margin of error. This tells us how confident we can be in our results.



6.   Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

=>  Skewness refers to the asymmetry in the distribution of data. It shows whether data values are spread more to one side of the mean than the other.

Positive Skew (Right-Skewed)
Tail is longer on the right (higher values)

Most data values are on the lower end

Mean > Median > Mode

 Example:
Income: Most people earn average or below average, but a few very high incomes pull the mean up.

2. Negative Skew (Left-Skewed)
Tail is longer on the left (lower values)

Most data values are on the higher end

Mean < Median < Mode

 Example:
Age of retirement: Most people retire around a common age, but a few retire very early, pulling the mean down.

3. No Skew (Symmetrical)
Data is evenly distributed

Mean = Median = Mode

 Example:
Heights of adults in a large population usually form a roughly symmetrical bell curve.

 How Skewness Affects Data Interpretation
 1. Affects the Mean
Skew pulls the mean toward the tail.

So the mean may not reflect the typical value if data is skewed.

 2. Influences Choice of Central Tendency
Median is better for skewed data, because it's less affected by outliers.

Mean is better for symmetrical data.

 3. Impacts Statistical Analysis
Many tests (e.g., t-tests, regression) assume normality. Skewness might violate that assumption, affecting results.

Sometimes, data transformation (like log or square root) is used to reduce skewness.

 4. Reveals Hidden Patterns
Skewness can indicate the presence of outliers or limitations in how data was collected or grouped.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

=>  A random variable is a numerical outcome of a random phenomenon or experiment. It assigns a number to each possible outcome of a random process.

Two Types of Random Variables
 1. Discrete Random Variables
Takes on a countable number of distinct values

Often represents things you can count

 Examples:
Number of students in a class

Number of heads in 10 coin flips

Number of cars passing through a toll booth in an hour

 Key Features:
Values can be listed: 0, 1, 2, 3, ...

Often modeled with binomial, Poisson, or geometric distributions

 2. Continuous Random Variables
Takes on an infinite number of possible values within a range

Represents things you can measure

 Examples:
Height of a person (e.g., 170.5 cm)

Time it takes to run a mile (e.g., 6.73 minutes)

Temperature in a room

 Key Features:
Values are not countable; they form a continuous interval

Probabilities are found over intervals, not individual values (since P(X = a) = 0)

Often modeled with normal, exponential, or uniform distributions
